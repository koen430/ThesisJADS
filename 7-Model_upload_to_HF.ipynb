```python
from google.colab import drive
import datetime
from datasets import (load_dataset,
                      Dataset,
                      DatasetDict)
import pandas as pd
import torch
from transformers import (AutoModelForCausalLM,
                          AutoTokenizer,
                          set_seed)
from peft import PeftModel
from tqdm.auto import tqdm
import accelerate
```

```python
drive.mount('/content/drive')
```

```python
# Set all the paths and parameters which can be changed for training
model = "microsoft/phi-2"
model_name = model.replace('.','').replace('-','_').replace('/','_')
source = "news" # twitter or news or mix
filestamp = "1711835043" # Select the timestamp of the latest version
timestamp = "1712334679" # Timestamp of the most recent trained adapter
checkpoint = "checkpoint-8800" # Checpoint with best performance

set_seed(42)
adapter_name = f"{timestamp}_{filestamp}_{source}_trained"

input_dir = f'/content/drive/MyDrive/Thesis_models/{adapter_name}/{checkpoint}'
```

```python
# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained(model,trust_remote_code=True,)
tokenizer.pad_token = tokenizer.eos_token
```

```python
# Load the base model
model = AutoModelForCausalLM.from_pretrained(
    model,
    device_map='auto',
)
```

```python
# Load the trained model
peft_model = PeftModel.from_pretrained(model,input_dir,is_trainable=False)
```

```python
peft_model.push_to_hub(f"koen430/Phi2_stockprediction_{source}")
```

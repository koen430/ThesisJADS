```python
# Function to nicely plot the loss
def get_loss_plot(data, title, timestamp):
  loss = data['loss'].dropna()
  eval_loss = data['eval_loss'].dropna()
  epochs = data['epoch'].drop_duplicates()[:len(eval_loss)]
  plt.figure(figsize=(12,5))
  plt.plot(epochs, loss)
  plt.plot(epochs, eval_loss)
  plt.legend(['training_loss', 'eval_loss'])
  plt.ylabel('loss')
  plt.xlabel('epochs')
  plt.grid()
  plt.title(title)
  plt.xticks(np.arange(1,max(epochs)+1))

  title = title.replace(' ', '_').replace(':','')
  plt.savefig(f'/content/drive/MyDrive/Thesis figures/{timestamp}_{title}')
  plt.show()
  print(f"Train time (s): {data['train_runtime'].dropna().values}")
```

# twitter training

5e-6, 10 epochs

```python
timestamp = "1712006432"
title = 'Causal twitter LR 5e-6'

twitter_logs = pd.read_csv(f'{path}/{timestamp}_{filestamp}_twitter')
get_loss_plot(twitter_logs, title, timestamp)
```

# News

```python
# Function to get nice loss plot
def get_loss_plot_news(data, title, timestamp):
  loss = data['loss'].dropna()
  eval_loss = data['eval_loss'].dropna()
  step = data['step'].drop_duplicates()[:len(eval_loss)]
  plt.figure(figsize=(12,5))
  plt.plot(step, loss)
  plt.plot(step, eval_loss)
  plt.legend(['training_loss', 'eval_loss'])
  plt.ylabel('loss')
  plt.xlabel('steps')
  plt.grid()
  plt.title(title)
  plt.xticks(np.arange(0,max(step)+800, min(step)))
  plt.xticks(rotation=90)

  title = title.replace(' ', '_').replace(':','')
  plt.savefig(f'/content/drive/MyDrive/Thesis figures/{timestamp}_{title}')
  plt.show()
  print(f"Train time (s): {data['train_runtime'].dropna().values}")
```

5e-6, 10 epochs

```python
timestamp = "1712334679"
title =  'Casual news LR: 5e-6'
news_logs = pd.read_csv(f'{path}/{timestamp}_{filestamp}_news')
get_loss_plot_news(news_logs, title, timestamp)
```
